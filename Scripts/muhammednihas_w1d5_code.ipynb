Data Cleaning

```python
print(df.duplicated().sum())
dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print(dupes.shape)
dupes.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['shipping Date'] = (df['Ship Date'] - df['Order Date']).dt.days
print(df['shipping Date'].describe())
```

```python
print(df.groupby('Region')[ 'shipping Date'].mean())
```

```python
df['Profit Margin'] = df['Profit'] / df['Sales']*100.
df['Profit Margin'] = df['Profit Margin'].round(2)
print(df.groupby('Sub-Category')['Profit Margin'].mean().sort_values().round(2))
```

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.rename(columns={'Sub-Category': 'sub_category', 'Ship Mode': 'ship_mode', 'Customer Name': 'customer_name',
                    'Order Year': 'order_year', 'shipping Date': 'shipping_date', 'Profit Margin': 'profit_margin', 
                    'Sales': 'sales', 'Profit': 'profit', 'Order Date': 'order_date', 'Ship Date': 'ship_date',
                    'Region': 'region', 'Customer ID': 'customer_id', 'Order ID': 'order_id', 'Category': 'category',
                    'Customer Name': 'customer_name', 'Row ID': 'row_id', 'Segment': 'segment', 'Quantity': 'quantity',
                    'Product Name': 'product_name','Discount': 'discount', 'Product ID': 'product_id',
                    'Country/Region': 'country_region','City': 'city', 'Postal Code': 'postal_code'}, inplace=True)
```
